```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def scrape_seek_jobs(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    job_titles = [title.text.strip() for title in soup.find_all('a', {'data-automation': 'jobTitle'})]
    company_names = [company.text.strip() for company in soup.find_all('a', {'data-automation': 'jobCompany'})]
    locations = [location.text.strip() for location in soup.find_all('a', {'data-automation': 'jobLocation'})]

    # Ensure all lists have the same length
    min_length = min(len(job_titles), len(company_names), len(locations))

    # Create DataFrame with equal-length lists
    jobs_data = pd.DataFrame({
        'Job Title': job_titles[:min_length],
        'Company': company_names[:min_length],
        'Location': locations[:min_length]
    })

    return jobs_data

# Example URL for data science jobs in Victoria, Australia on Seek.com.au
seek_url = 'https://www.seek.com.au/Data-Analyst-jobs/in-All-Melbourne-VIC'

# Scrape job data
jobs_data = scrape_seek_jobs(seek_url)

# Display the first few rows of the data
jobs_data.head()
```
